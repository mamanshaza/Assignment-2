```python
# compare average monthly income by Education and Attrition

Attr_new2 = ATTR.iloc[:,[1,6,19]]
pd.set_option('display.max_rows',None)
Attr_new2
```

```python
Education = Attr_new2.groupby("Attrition")["Education"]
print(Education.value_counts())
```

```python
MonthlyRate = Attr_new2.groupby("Attrition")["MonthlyRate"]
print(MonthlyRate.value_counts())

sns.displot(data=Attr_new2, x="MonthlyRate", hue = "Attrition", height = 10)
```

```python
sns.displot(data=Attr_new2, x="MonthlyRate", hue = "Attrition", height = 10)
```

```python
Attr_numerized = Attr_new2


for col_name in Attr_numerized.columns:
    if(Attr_numerized[col_name].dtype == 'object'):
        Attr_numerized[col_name]= Attr_numerized[col_name].astype('category')
        Attr_numerized[col_name] = Attr_numerized[col_name].cat.codes
        
Attr_numerized
```

```python
sns.displot(x="MonthlyRate", y="Education", hue= 'Attrition', data=Attr_numerized)
```

```python
sns.displot(x="Education", y="MonthlyRate", hue='Attrition', data=Attr_numerized)
```

```python
# Correlation Matrix between all numeric columns

Attr_numerized.corr(method ='pearson')
```

```python
Attr_numerized.corr(method ='kendall')
```

```python
Attr_numerized.corr(method ='spearman')
```

```python
correlation_matrix = Attr_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for MonthlyRate")

plt.xlabel("MonthlyRate features")

plt.ylabel("MonthlyRate features")

plt.show()

```
